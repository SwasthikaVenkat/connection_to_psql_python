```python
import psycopg2
```

```python
#establishing the connection
conn = psycopg2.connect(
   database="postgres", user='postgres', password='swasthika', host='localhost', port= '5432'
)
```

```python
#Creating a cursor object using the cursor() method
cur = conn.cursor()
```

```python
#Executing an MYSQL function using the execute() method
cur.execute("SELECT * FROM employees")
#print(cur.fetchall())
for row in cur.fetchall():
    print(row)
```

```python
import pandas as pd
pd.set_option("display.max_rows",None)
my_table    = pd.read_sql('select * from employees', conn)
my_table
```

```python
cur.execute('DROP TABLE IF EXISTS sales')
```

```python
create_script = ''' CREATE TABLE IF NOT EXISTS sales (
                                    id      int PRIMARY KEY,
                                    name    varchar(40) NOT NULL,
                                    salary  int,
                                    dept varchar(30)) '''
cur.execute(create_script)
```

```python
insert_script  = 'INSERT INTO sales (id, name, salary, dept) VALUES (%s, %s, %s, %s)'
insert_values = [(1, 'James', 12000, 'cleaning'), (2, 'Robin', 15000, 'cashier'), (3, 'Xavier', 20000, 'accounts')]
for record in insert_values:
    cur.execute(insert_script, record)
```
